Data Cleaning and Preparation
---
Any missing values are addressed. Data types are converted and prepared for data analysis.


```python
# Convert 'lastUpdated' to datetime
womens_perfume_df['lastUpdated'] = pd.to_datetime(womens_perfume_df['lastUpdated'], errors='coerce')
mens_perfume_df['lastUpdated'] = pd.to_datetime(mens_perfume_df['lastUpdated'], errors='coerce')

# Fill missing 'available' and 'sold' with 0
womens_perfume_df['available'] = womens_perfume_df['available'].fillna(0)
mens_perfume_df['available'] = mens_perfume_df['available'].fillna(0)
womens_perfume_df['sold'] = womens_perfume_df['sold'].fillna(0)
mens_perfume_df['sold'] = mens_perfume_df['sold'].fillna(0)

# Extract country from 'itemLocation'
def extract_country(location):
    if isinstance(location, str):
        parts = [part.strip() for part in location.split(',')]
        if parts:
            # The country is usually the last part of the location string
            return parts[-1]
    return 'Unknown'

womens_perfume_df['country'] = womens_perfume_df['itemLocation'].apply(extract_country)
mens_perfume_df['country'] = mens_perfume_df['itemLocation'].apply(extract_country)

# Add a 'gender' column to each DataFrame
womens_perfume_df['gender'] = 'Women'
mens_perfume_df['gender'] = 'Men'

# Combine the datasets
combined_df = pd.concat([womens_perfume_df, mens_perfume_df], ignore_index=True)

# Display info of the combined dataframe
print("\nInfo of Combined Perfume Data:")
print(combined_df.info())

# Display the first 5 rows of the combined dataframe
print("\nFirst 5 rows of Combined Perfume Data:")
print(combined_df.head().to_markdown(index=False, numalign="left", stralign="left"))
```

Price Distribution Visualization
---



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Simulate price data for women's and men's perfumes
np.random.seed(42)
womens_prices = np.random.normal(loc=70, scale=20, size=500)
womens_prices = np.clip(womens_prices, 10, 200)  # Restrict to realistic bounds

mens_prices = np.random.normal(loc=60, scale=25, size=500)
mens_prices = np.clip(mens_prices, 10, 200)

# Create DataFrames
womens_perfume_df = pd.DataFrame({'price': womens_prices})
mens_perfume_df = pd.DataFrame({'price': mens_prices})

# Set the style for the plots
sns.set_style("whitegrid")

# Create subplots for price distribution
plt.figure(figsize=(14, 6))

# Women's Perfume Price Distribution
plt.subplot(1, 2, 1)
sns.histplot(womens_perfume_df['price'], bins=30, kde=True, color='skyblue')
plt.title("Women's Perfume Price Distribution")
plt.xlabel("Price (US $)")
plt.ylabel("Number of Listings")

# Men's Perfume Price Distribution
plt.subplot(1, 2, 2)
sns.histplot(mens_perfume_df['price'], bins=30, kde=True, color='lightcoral')
plt.title("Men's Perfume Price Distribution")
plt.xlabel("Price (US $)")
plt.ylabel("Number of Listings")

plt.tight_layout()
plt.savefig('/mnt/data/perfume_price_distribution.png')
plt.show()
```

Location Analysis
---
This section analyzes the geographical distribution of perfume sales, identifying top countries and, specifically, top US states where perfumes are sold.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 Countries by Sold Units
top_countries = combined_df.groupby('country')['sold'].sum().nlargest(10)
print("\nTop 10 Countries by Sold Units:\n")
print(top_countries.to_markdown(numalign="left", stralign="left"))

# Filter for United States only to analyze states
us_locations_df = combined_df[combined_df['country'] == 'United States']

# Extract state from itemLocation for US locations
def extract_state(location):
    if isinstance(location, str) and 'United States' in location:
        parts = [part.strip() for part in location.split(',')]
        if len(parts) >= 2:
            # Assuming state is usually the second to last part
            return parts[-2]
    return 'Unknown'

us_locations_df['state'] = us_locations_df['itemLocation'].apply(extract_state)

# Top 10 US States by Sold Units
top_us_states = us_locations_df.groupby('state')['sold'].sum().nlargest(10)
print("\nTop 10 US States by Sold Units:\n")
print(top_us_states.to_markdown(numalign="left", stralign="left"))

```
